```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
json_path = "./Hierarchical-Localization/cam_info.json"
with open(json_path, 'r') as f:
    cam_info = json.load(f)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax_2d = fig.add_subplot()

images = cam_info["images"]
locs  = cam_info["loc"]
fronts = cam_info["front"]
print(cam_info)
for i in range(len(images)):
    if "F" in images[i]:
        print(locs[i], fronts[i])

        ax.scatter(*locs[i], c="red")
        ax.scatter(*fronts[i], c="blue")
        ax.set_xlim(-5,5)
        ax.set_ylim(-5,5)
        ax.set_zlim(-5,5)
        
        ax_2d.scatter(locs[i][0], locs[i][2], c="red")
        ax_2d.scatter(fronts[i][0], fronts[i][2], c="blue")
        ax_2d.annotate(images[i], (locs[i][0], locs[i][2]))
        ax_2d.set_xlim(-6,6)
        ax_2d.set_ylim(-6,6)
```
